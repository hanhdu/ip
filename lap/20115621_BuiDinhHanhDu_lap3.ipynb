```python
#Khai báo thư viện
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def histogram_equalization (image):
    #Tính toán hitogram
    hist, _ = np.histogram (image.flatten(), 256, [0, 256])
    
    #Tính toán hàm tích lũy 
    cdf = hist.cumsum()
   
    #Chuẩn hóa hàm tích lũy 
    cdf_normarlized = cdf * 255 / cdf[-1]
   
    #Ánh xạ lại các mức xám theo hàm tich lũy chuẩnn hóa 
    equalized_image = cdf_normarlized[image]

    return equalized_image.astype(np.uint8)
```

```python
#Đọc hình ảnh
img = cv2.imread('img.png', cv2.IMREAD_GRAYSCALE)

#Thực hiện cân bằng histogram
histogram_image = histogram_equalization(img)

#sử dụng thư viện cân bằng histogram
histogram_image_li = cv2.equalizeHist(img)
```

```python
#show image
plt.figure(figsize= (15, 5))
plt.subplot(131),plt.imshow(img, cmap= 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(histogram_image, cmap= 'gray'), plt.title('Histogram equalization')
plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(histogram_image_li, cmap= 'gray'), plt.title('Histogram equalization dùng thư viện')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

def mean_filter(image, kernel_size):

    padded_image = cv2.copyMakeBorder (image, kernel_size//2, kernel_size//2, kernel_size//2, kernel_size//2, 
                                      cv2.BORDER_CONSTANT)
    output_image = np.zeros_like(image)

    for i in range(image.shape[0]):
        for j in range (image.shape[1]):
            neigborhood = padded_image[i:i +kernel_size, j: j +kernel_size]
            mean_vallue = np.mean(neigborhood)
            output_image[i, j] = mean_vallue
    return output_image
```

```python
mean_image = mean_filter(img, 7)

kernel_size = (7 ,7)
mean_image_li = cv2.blur(img, kernel_size)
```

```python
plt.figure(figsize= (15, 5))
plt.subplot(131),plt.imshow(img, cmap= 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(mean_image, cmap= 'gray'), plt.title('Mean filter')
plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(mean_image_li, cmap= 'gray'), plt.title('Maen filter dùng thư viện')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def median_filter(image, kernel_size):
    h, w = image.shape
    filtered_image = np.zeros((h, w),dtype=np.uint8)

    window_size = kernel_size*kernel_size

    radius = kernel_size//2

    for i in range (radius, h - radius):
        for j in range (radius, w - radius):
            window = image[i - radius: i + radius + 1, j - radius: j + radius + 1]
            median_value = np.median(window)
            filtered_image[i, j] = median_value    

    return filtered_image
```

```python
median_image = median_filter(img, 5)

median_image_li = cv2.medianBlur(img, 5)
```

```python
plt.figure(figsize= (15, 5))
plt.subplot(131),plt.imshow(img, cmap= 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(median_image, cmap= 'gray'), plt.title('Median filter')
plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(median_image_li, cmap= 'gray'), plt.title('Median filter dùng thư viện')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def min_filter(image, kernel_size):
    h, w = image.shape
    filtered_image = np.zeros((h, w),dtype=np.uint8)

    window_size = kernel_size*kernel_size

    radius = kernel_size//2

    for i in range (radius, h - radius):
        for j in range (radius, w - radius):
            window = image[i - radius: i + radius + 1, j - radius: j + radius + 1]
            min_value = np.min(window)
            filtered_image[i, j] = min_value    

    return filtered_image
```

```python
min_image = min_filter(img, 5)

min_image_li = cv2.filter2D(img,-1, (5, 5))
```

```python
plt.figure(figsize= (15, 5))
plt.subplot(131),plt.imshow(img, cmap= 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(min_image, cmap= 'gray'), plt.title('Min filter')
plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(min_image_li, cmap= 'gray'), plt.title('Min filter dùng thư viện')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def max_filter(image, kernel_size):
    h, w = image.shape
    filtered_image = np.zeros((h, w),dtype=np.uint8)

    window_size = kernel_size*kernel_size

    radius = kernel_size//2

    for i in range (radius, h - radius):
        for j in range (radius, w - radius):
            window = image[i - radius: i + radius + 1, j - radius: j + radius + 1]
            max_value = np.max(window)
            filtered_image[i, j] = max_value    

    return filtered_image
```
