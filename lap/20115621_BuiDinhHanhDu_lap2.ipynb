```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Get the dimensions of the image
height, width = img.shape

# Calculate the length of the left and right borders
border_width = int(width * 0.1)

# Calculate the length of the upper and lower borders
border_height = int((width + 2 * border_width - height) / 2)

# Create a black border around the image
border = cv2.copyMakeBorder(img, border_height, border_height, border_width, border_width, cv2.BORDER_CONSTANT, None, (0, 0, 0))

#show image
plt.subplot(121),plt.imshow(img, cmap= 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(border, cmap= 'gray'), plt.title('Img_Border')
plt.xticks([]), plt.yticks([])
plt.show()

# Save the image
cv2.imwrite('image_border.jpg', border)
```

```python

# Load the image in grayscale
img = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

# Rotate the image by 90 degrees
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

# Show image 
plt.subplot(121),plt.imshow(img, cmap= 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(rotated, cmap= 'gray'), plt.title('Rotated 90')
plt.xticks([]), plt.yticks([])
plt.show()

# Save the rotated image
cv2.imwrite('rotated_image.jpg', rotated)
```

```python
def number_white_objects (image):
    # Load the image
    img = cv2.imread(image, cv2.IMREAD_GRAYSCALE)

    # Threshold the image
    _, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

    # Apply connected component labelling
    num_labels, labels = cv2.connectedComponents(thresh)

    # Count the total number of white objects
    num_white_objects = num_labels - 1

    return num_white_objects


print('Total number of white objects in photo a:', number_white_objects('image1.jpg'))
print('Total number of white objects in photo b:', number_white_objects('image2.jpg'))
print('Total number of white objects in photo c:', number_white_objects('image3.jpg'))
```
