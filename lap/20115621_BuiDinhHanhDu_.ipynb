```python
#khái báo thư viện
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

1. Đọc ảnh và chuyển về ảnh Gray-scale

```python
#Đọc ảnh và chuyển ảnh sang màu xám
img = cv2.imread('input.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap= 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.show()
```

2. Vẽ lược đồ histogram

```python
histogram = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.figure()
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```

3. Thực hiện cân bằng Histogram

```python
# Cân bằng histogram
equalized_image = cv2.equalizeHist(img)

# Hiển thị ảnh gốc và ảnh đã cân bằng histogram
plt.figure()
plt.subplot(121)
plt.imshow(img, cmap= 'gray'),plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(equalized_image, cmap='gray'), plt.title('Equalized Hitogram Image')
plt.xticks([]), plt.yticks([])

plt.show()
```

4. Resize ảnh về kích thước 300x300

```python
resized_image = cv2.resize(img, (300, 300))

plt.imshow(resized_image, cmap= 'gray'),plt.title('Image Resized')

plt.show()
```

5. Thực hiện tính đạo hàm theo hướng x và y sử dụng mặt nạ. 

MSSV:20115621 -> Prewitt

```python
# Apply the Prewitt filter
prewitt_x = np.array([[-1, 0, 1],
                         [-1, 0, 1],
                         [-1, 0, 1]])
prewitt_y = np.array([[-1, -1, -1],
                         [0, 0, 0],
                         [1, 1, 1]])

gradient_x = cv2.filter2D(img, -1, prewitt_x)
gradient_y = cv2.filter2D(img, -1, prewitt_y)

#Hiển thị kết quả
plt.figure(figsize= (9, 9))

plt.subplot(221),plt.imshow(img, cmap= 'gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])


plt.subplot(223),plt.imshow(gradient_x, cmap= 'gray'), plt.title('gradient_x')
plt.xticks([]), plt.yticks([])

plt.subplot(224),plt.imshow(gradient_y, cmap= 'gray'), plt.title('gradient_y')
plt.xticks([]), plt.yticks([])


plt.show()
```

6. Tính cường độ Gradient

```python
gradient_magnitude = gradient_x + gradient_y

plt.figure()
plt.subplot(121)
plt.imshow(img, cmap= 'gray'),plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Gradient Magnitude')
plt.xticks([]), plt.yticks([])

plt.show()
```

7. Phân đoạn ảnh bằng ngưỡng Otsu

```python
_, segmented_image = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(9, 8))
plt.subplot(121)
plt.imshow(img, cmap= 'gray'),plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(segmented_image, cmap='gray'), plt.title('Images segmented otsu thresholding method')
plt.xticks([]), plt.yticks([])

plt.show()

```
